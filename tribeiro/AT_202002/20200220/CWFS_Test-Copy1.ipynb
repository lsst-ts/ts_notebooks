```python
import sys
import asyncio
import logging

import numpy as np

from lsst.ts import salobj

import matplotlib.pyplot as plt

from lsst.ts.externalscripts.auxtel.latiss_cwfs_align import LatissCWFSAlign
```

```python
stream_handler = logging.StreamHandler(sys.stdout)

logger = logging.getLogger()
logger.addHandler(stream_handler)
logger.level = logging.DEBUG
```

```python
cwfs = LatissCWFSAlign(index=1)
```

```python
await cwfs.start_task
```

```python
await cwfs.attcs.slew_object("HD 116351", pa_ang=0.)
```

```python
await cwfs.latiss.take_object(exptime=1,n=1, filter="BG40", grating = "empty_1")
```

```python
cwfs.intra_visit_id = 2020022000337
cwfs.extra_visit_id = 2020022000338
cwfs.angle = 0.
# # print(cwfs.camera_rotation_angle)
```
