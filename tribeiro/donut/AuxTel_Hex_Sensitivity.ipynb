```python
%matplotlib inline
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def parabola(x,b, x0, a):
    return b + a*(x-x0)**2 
def line(x,b, x0, a):
    return b + a*(x-x0) 
```
