```python
import sys
import asyncio
import time
import os
import numpy as np

from lsst.ts import salobj
from lsst.ts.observatory.control.maintel.comcam import ComCam
```

```python
import logging
stream_handler = logging.StreamHandler(sys.stdout)
logger = logging.getLogger()
logger.addHandler(stream_handler)
logger.level = logging.DEBUG
# Make matplotlib less chatty
logging.getLogger("matplotlib").setLevel(logging.WARNING)
```

```python
#Start classes
domain = salobj.Domain()
comcam = ComCam(domain)
await comcam.start_task
```

```python
# enable components
await comcam.enable()
```

```python
# Take a bias to make sure everything is working
await comcam.take_bias(1)
```

```python
# Take 50 biases
# Added wait to stop killing the recent images
for i in range(50):
    await asyncio.sleep(2.0)
    await comcam.take_bias(1)
```

```python
# Take 10 100 second darks 
await comcam.take_darks(100.0, 10)
```

```python
# I haven't tried this, but it should work
await comcam.setup_filter('i_06')
```

```python
# Take flats for PTC
# Added wait to stop killing the recent images
maxExpTime = 5.0 # Set this to get enough signal.
for i in range(20):
    exp = maxExpTime / 20.0 * float(i+1)
    await comcam.take_flats(exp, 1)
    if exp < 2.0:
        await asyncio.sleep(2.0)
    await comcam.take_flats(exp, 1)
```
