# Looking for interference
There's a suspicion that the TMA motors may be causing signal interference inside of the dome network. There is a Spectrum Analyzer on the 7th floor that may be able to detect this interference should it exist. This notebooks covers efforts to check the data.

```python
from lsst_efd_client import EfdClient
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
from astropy.time import Time
```

```python
efd_client = EfdClient(efd_name="summit_efd")
```

```python
fields = []
for spectrum in range(751):
    fields.append(f"spectrum{spectrum}")
```

We'll start with taking a minute worth of spectra before the TMA is logged to have moved.

```python
spectrum_data = await efd_client.select_time_series(
    topic_name="lsst.sal.ESS.spectrumAnalyzer",
    fields=fields,
    start=Time("2023-06-28T18:33:00"),
    end=Time("2023-06-28T18:34"))
```

```python
spectrum_data.describe()
```

Let's check the spectra during motion.

```python
motion_spectrum_data = await efd_client.select_time_series(
    topic_name="lsst.sal.ESS.spectrumAnalyzer",
    fields=fields,
    start=Time("2023-06-28T18:34:15"),
    end=Time("2023-06-28T18:37:00"))
```

```python
motion_spectrum_data.describe()
```

```python
fields = []
for current in range(16):
    fields.append(f"current{current}")
```

Let's look at the elevation drives current values.

```python
elevation_current_data = await efd_client.select_time_series(topic_name="lsst.sal.MTMount.elevationDrives",fields=fields, start=Time("2023-06-28T18:34:15"), end=Time("2023-06-28T18:37:16"))
```

```python
elevation_current_data.describe()
```

It doesn't seem like there are huge variations in the spectra when TMA is recorded as moving and when it is recorded as not moving.
