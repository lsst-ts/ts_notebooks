```python
#Start classes
domain = salobj.Domain()
await asyncio.sleep(10) # This can be removed in the future...
atcs = ATCS(domain)
latiss = LATISS(domain)
await asyncio.gather(atcs.start_task, latiss.start_task)
```

```python
# enable components, then take 50 more biases
await atcs.enable()
await latiss.enable()
```

```python
# All components now enabled. - 02-Feb-2021 12:54 PM
```

```python
await salobj.set_summary_state(atcs.rem.atdometrajectory, salobj.State.DISABLED)
```

```python
# I think these will prevent the dome from moving.
# Take event checking out the slew commands to test telescope only
atcs.check.atdome = False
atcs.check.atdometrajectory = False
```

```python
await atcs.slew_icrs(ra="22:00:00", dec="-20:00:00", rot=0.0)
```

```python
await atcs.slew_icrs(ra="22:00:00", dec="-40:00:00", rot=0.0)
```

```python
# Now run 50 biases, separated by 15 second delay - Images 354-403
for i in range(50):
    await latiss.take_bias(1)
    await asyncio.sleep(15)
```

```python
# Move telescope slightly from park position (az=0, el=80, rot=0)
current_az = 5.0
current_el = 75.0
current_rot = 10.0
await atcs.point_azel(current_az, current_el, rot_tel=current_rot)
```
