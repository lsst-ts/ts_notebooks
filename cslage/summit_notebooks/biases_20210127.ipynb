```python
import sys
import asyncio
import time
import os

import numpy as np

from lsst.ts import salobj

from lsst.ts.observatory.control.auxtel.atcs import ATCS
from lsst.ts.observatory.control.auxtel.latiss import LATISS
```

```python
import logging
stream_handler = logging.StreamHandler(sys.stdout)
logger = logging.getLogger()
logger.addHandler(stream_handler)
logger.level = logging.DEBUG
```

```python
#Start LATISS class
domain = salobj.Domain()
await asyncio.sleep(10) # This can be removed in the future...
atcs = ATCS(domain)
latiss = LATISS(domain)
await asyncio.gather(atcs.start_task, latiss.start_task)
```

```python
await salobj.set_summary_state(latiss.rem.atarchiver, salobj.State.ENABLED)
```

```python
await salobj.set_summary_state(latiss.rem.atheaderservice, salobj.State.ENABLED)
```

```python
# Run 1 bias as a test
await latiss.take_bias(1)
```

```python
# Now run 1000, separated by 15 second delay
# First will be 006
for i in range(1000):
    await latiss.take_bias(1)
    await asyncio.sleep(15)
```
