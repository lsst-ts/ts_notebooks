## Query header

In this notebook, we show several ways to query the fits headers\

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as pf
from lsst.daf.persistence import Butler
import lsst.afw.cameraGeom.utils as camGeomUtils
import lsst.afw.display as afwDisplay
```

```python
REPO_DIR = '/project/shared/auxTel'
butler = Butler(REPO_DIR)
dayObs = '2021-06-10'
```

```python
# This queries the header information in the butler metadata
# This is a limited amount of information, but returns very quickly
visits = butler.queryMetadata('raw', ['expId', 'EXPTIME', 'imagetype', 'object', 'filter', 'DATE'],\
                              detector=0, dayObs=dayObs)
for visit in visits:
    print(visit)
```

```python
# This returns all of the information in the header, but takes longer
expId=2021060800179
exp = butler.get('raw', detector=0, expId=expId)
mData = exp.getMetadata()
for key in mData.keys():
    print(key, mData[key])
```

```python
# This way goes directly to the fits files without using the butler
import glob
files = glob.glob(REPO_DIR+'/_parent/raw/2021-03-23/*')
print(len(files))
hduList = pf.open(files[320])
hdr = hduList[0].header
for key in hdr.keys():
    print(key, hdr[key])
```
