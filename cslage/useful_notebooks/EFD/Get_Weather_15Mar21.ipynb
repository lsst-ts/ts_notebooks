```python
import sys, time, os, asyncio

from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.time import Time, TimeDelta
from lsst_efd_client import EfdClient
from lsst.daf.persistence import Butler
```

```python
# Get EFD client and the butler
client = EfdClient('summit_efd')
REPO_DIR = '/project/shared/auxTel/rerun/quickLook'
butler = Butler(REPO_DIR)
```

```python
# Find the time of exposure
expId=2021031100348
tStart = butler.queryMetadata('raw', ['DATE'], detector=0, expId=expId)# Get the data
t_start = Time(tStart, scale='tai')[0]
t_end = t_start + TimeDelta(60, format='sec')
```

```python
wind_speed = await client.select_time_series("lsst.sal.WeatherStation.windSpeed", ['*'],
                                          t_start, t_end)
wind_direction = await client.select_time_series("lsst.sal.WeatherStation.windDirection", ['*'],
                                          t_start, t_end)
```
