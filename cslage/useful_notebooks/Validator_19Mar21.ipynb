```python
import yaml
from lsst.ts import salobj
```

```python
schema_file = "/home/craiglagegit/WORK/ts_atspec/schema/ATSpectrograph.yaml"
with open(schema_file, "r") as f:
     schema = yaml.safe_load(f.read())
validator = salobj.DefaultingValidator(schema)
config_path = "/home/craiglagegit/WORK/ts_config_latiss/ATSpectrograph/v3/LATISS_AuxTel_20210318.yaml"
with open(config_path, "r") as f:
     config_data = yaml.safe_load(f.read())
validator.validate(config_data)
```
